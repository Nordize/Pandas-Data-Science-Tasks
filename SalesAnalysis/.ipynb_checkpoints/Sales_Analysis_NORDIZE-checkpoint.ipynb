#### Task #1: Merging 12 months of Sales data into a single csv file 

```python
#df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")  #read one file

#read all file from directories python
files = [file for file in os.listdir('./Sales_Data/')]

all_months_data = pd.DataFrame() #create an empty dataFrame

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data.to_csv("./Merged_file/all_data.csv",index=False)
```

#### Question 2: What city had the highest number of sales

#### Task 4: Add a city column

```python
all_data.head()
```

```python
# to split a month, we use .apply()
#all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
#all_data.head()

#Function way
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

#all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + " " + get_state(x))
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")  #use f-string (python 3.5)
all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]  #Switching city axis to x 

plt.bar(cities,results['Sales'])

plt.xticks(cities,rotation = 'vertical',size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Cities')
plt.show()
```

#### ANS: San Fran has the highest number of sales due to the population, shops and salary

#### Question 4: What products are most often sold together

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]  #will check or the row or cell in the order_id column that duplicate

df.head(20)
```

```python
#create new column called Grouped

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
#drop duplicate rows

df = df[['Order ID','Grouped']].drop_duplicates()

df.head()
```

```python
# reference https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))  # match those 2 items that sell together
    #count.update(Counter(combinations(row_list,3)))  # match those 3 items that sell together
    
for key, value in count.most_common(10):
    print(key,value)
```

#### Ans: 'iPhone', 'Lightning Charging Cable' are those items that sell together, might be do ads with them together is a good idea

#### Question 5: What product sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
product_group.sum()
```

```python
#plot

quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products,quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.xticks(products,rotation = 'vertical',size = 8)
plt.show()
```

```python
#to see more detail
prices = all_data.groupby('Product').mean()['Price Each']

print(prices)
```

```python
#plot second y-label

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products,quantity_ordered,color='g')
ax2.plot(products,prices,'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered',color='g')
ax2.set_ylabel('Price ($)',color = 'b')
ax1.set_xticklabels(products,rotation = 'vertical',size = 8)

plt.show()
```

#### ANS: we can see that both AA and AAA batteries are sold the most and also both charging cable, might be because its cheap, easy to lose and everyone need it. And whenever the quantity is hight price should be low which is the case for both cable and batteries. Also happens with the laptop which is price is high then quantity is low.
